## Dataset Source and Columns

The dataset used is the **Apple Stock Market Dataset**, which contains historical stock price data for Apple Inc.

Source: Kaggle – Apple Historical Stock Prices Dataset

The dataset includes the following columns:

- Date – Trading date
- Open – Opening stock price
- High – Highest price of the day
- Low – Lowest price of the day
- Close – Closing price
- Adj Close – Adjusted closing price
- Volume – Number of shares traded

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("seaborn-v0_8")
sns.set_palette("Set2")
```

```python
df = pd.read_csv("apple_stock.csv")

print("Initial Shape:", df.shape)
print("\nFirst 5 Rows:\n", df.head())
```

```python
print("\nDuplicate Rows:", df.duplicated().sum())
print("\nMissing Values:\n", df.isnull().sum())

# Rename date column
df.rename(columns={"Unnamed: 0": "Date"}, inplace=True)

# Convert to datetime
df["Date"] = pd.to_datetime(df["Date"])

# Sort by date
df = df.sort_values("Date")

# Remove duplicates if any
df = df.drop_duplicates()

print("\nFinal Shape After Cleaning:", df.shape)
```

```python
print("\nStatistical Summary:\n", df.describe())

print("\nMean Closing Price:", round(df["Adj Close"].mean(),2))
print("Median Closing Price:", round(df["Adj Close"].median(),2))
print("Standard Deviation:", round(df["Adj Close"].std(),2))
```

## Statistical Summary of Stock Prices

A statistical analysis was conducted to understand the distribution of Apple stock prices across the dataset.

The dataset contains **11,107 trading records** spanning multiple decades of stock market activity.

Key statistical observations include:

- The **average adjusted closing price is approximately 23.52**, reflecting the overall long-term price level.
- The **median price is 0.45**, indicating that earlier historical prices were significantly lower before Apple's major growth periods.
- The **maximum recorded price reaches 259.02**, highlighting the strong growth of Apple stock over time.
- The **minimum price is approximately 0.037**, reflecting the very early trading years of the company.
- The **standard deviation of 49.77** indicates noticeable price volatility across the time period.

These statistics highlight the dramatic long-term growth and variability in Apple stock prices.

```python
#Stock Price Trend Analysis
plt.figure(figsize=(12,6))
plt.plot(df["Date"], df["Adj Close"])
plt.title("Apple Adjusted Closing Price Over Time")
plt.xlabel("Date")
plt.ylabel("Adjusted Close Price")
plt.show()
```

## Stock Price Trend Analysis

The historical stock price trend shows a clear **long-term upward trajectory** in Apple’s adjusted closing price.

During the early years of trading, prices remained relatively low. However, over time the stock experienced substantial growth, particularly in the later years of the dataset.

This upward trend reflects Apple’s strong business expansion, innovation in consumer technology, and increasing investor confidence in the company’s performance.

```python
#Moving Average Trend Analysis
df["MA_50"] = df["Adj Close"].rolling(window=50).mean()
df["MA_200"] = df["Adj Close"].rolling(window=200).mean()

plt.figure(figsize=(12,6))
plt.plot(df["Date"], df["Adj Close"], label="Adj Close")
plt.plot(df["Date"], df["MA_50"], label="50-Day MA")
plt.plot(df["Date"], df["MA_200"], label="200-Day MA")
plt.legend()
plt.title("Moving Averages Analysis")
plt.show()
```

## Moving Average Trend Analysis

Moving averages were used to analyze long-term price trends.

Two common indicators were applied:

- **50-day moving average**
- **200-day moving average**

When the 50-day moving average crosses above the 200-day moving average, it often signals a **bullish trend**, suggesting potential long-term price growth.

These moving averages help smooth short-term fluctuations and provide clearer insight into long-term market direction.

```python
##Daily Return Analysis
df["Daily_Return"] = df["Adj Close"].pct_change()

print("\nAverage Daily Return:", round(df["Daily_Return"].mean(),5))
print("Daily Return Std Dev (Volatility):", round(df["Daily_Return"].std(),5))

plt.figure(figsize=(10,5))
sns.histplot(df["Daily_Return"].dropna(), bins=50, kde=True)
plt.title("Daily Return Distribution")
plt.show()
```

## Daily Return Analysis

Daily returns were calculated to evaluate short-term price movements in Apple stock.

Most daily returns remain within a **narrow range around zero**, indicating that small price fluctuations are common in daily trading activity.

However, occasional extreme movements are also observed, reflecting market reactions to major economic events, earnings announcements, or global financial changes.

Analyzing daily returns helps investors understand the risk and variability associated with short-term trading.

```python
#Volatility Analysis
df["Rolling_Volatility"] = df["Daily_Return"].rolling(window=30).std()

plt.figure(figsize=(12,6))
plt.plot(df["Date"], df["Rolling_Volatility"])
plt.title("30-Day Rolling Volatility")
plt.show()
```

## Volatility Analysis

Volatility measures the degree of variation in stock prices over time.

The standard deviation of daily returns provides an estimate of this volatility. In this dataset, Apple stock demonstrates **moderate volatility**, meaning that price changes occur regularly but generally remain within a manageable range.

Understanding volatility is important for investors because higher volatility indicates greater investment risk.

```python
#Trading Volume Analysis
plt.figure(figsize=(12,5))
plt.plot(df["Date"], df["Volume"])
plt.title("Trading Volume Over Time")
plt.show()
```

## Trading Volume Analysis

Trading volume represents the number of shares traded during a given time period.

The analysis shows that trading volume fluctuates over time, with occasional spikes during periods of significant market activity.

High trading volume often occurs during major events such as earnings announcements, product launches, or broader market changes.
